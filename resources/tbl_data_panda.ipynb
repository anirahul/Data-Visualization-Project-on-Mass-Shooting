```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file = "data/data.csv"

# Read Purchasing File and store into Pandas data frame
data = pd.read_csv(file)

data.head()
```

```python
data.columns
```

```python

data_df = pd.DataFrame({'Date':data['full_date'], 'City':data['city'], 'State': data['state'], 'Location':data['location'],'Gender': data['gender'],
                       'Age':data['age'], 'Killed':data['number_killed'], 'Weapons':data['total_weapons_brought_to_the_scene'], 'Crime/Violence Factor':data['crime_violence_confact_pct'],
                        'Trauma/Crisis Factor':data['trauma_crisis_confact_pct'], 'Mental Health Factor':data['mental_health_confact_pct'],
                        'Motivation Factor':data['motivation_confact_pct'], 'Social Factor':data['social_contagion_confact_pct']})



data2=data_df.sort_values(["Killed"],ascending=False)

data2=data2.reset_index(drop=True,inplace=False)

data2.head()
```

```python
#render dataframe as html
html = data2.to_html()

```

```python
#write html to file
text_file = open("DataRaw.html", "w")
text_file.write(html)
text_file.close()
```
